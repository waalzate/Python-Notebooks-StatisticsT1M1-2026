```python
# Install dependencies as needed:
# pip install kagglehub[pandas-datasets]
import kagglehub
from kagglehub import KaggleDatasetAdapter

# Set the path to the file you'd like to load
file_path = "exams.csv"

# Load the latest version
df = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "rkiattisak/student-performance-in-mathematics",
  file_path,
  # Provide any additional arguments like
  # sql_query or pandas_kwargs. See the
  # documenation for more information:
  # https://github.com/Kaggle/kagglehub/blob/main/README.md#kaggledatasetadapterpandas
)

print("First 5 records:", df.head())
```

```python
df
```

```python
## Filtramos la base de datos para el nivel de educación some college

df_some_college = df[df['parental level of education'] == 'some college']
```

```python
df_some_college
```

```python
df['parental level of education'].value_counts()
```

```python
## Filtramos la base de datos para el nivel de educación master degree


df_master_degree = df[df['parental level of education'] ==  "master's degree"]
```

```python
df_master_degree
```

```python
import numpy as np
```

```python
mean_level_some_college = np.mean(df_some_college['reading score'])
mean_level_master_degree = np.mean(df_master_degree['reading score'])
```

```python
print(mean_level_some_college,mean_level_master_degree  )
```

```python
# Calculemos las desviaciones estandar

#df_master_degree

#df_some_college

desv_some_college = np.std(df_some_college['reading score'])
desv_master_degree = np.std(df_master_degree['reading score'])

print(desv_some_college, desv_master_degree)
```

```python
# Quantiles

# Quantiles para la variable reading score y nivel de educación correspondiente a some college

quantiles_some_college = np.quantile(df_some_college['reading score'], [0, 0.25, 0.5, 0.75, 1])
print(quantiles_some_college )

# Quantiles para la variable reading score y nivel de educación correspondiente a master degree

quantiles_master_degree = np.quantile(df_master_degree['reading score'], [0, 0.25, 0.5, 0.75, 1])
print(quantiles_master_degree )
```
